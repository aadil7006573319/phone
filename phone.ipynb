1. What is a null hypothesis (H₀) and why is it important in hypothesis testing?

  -A **null hypothesis (H₀)** is a statement in statistics that assumes there is **no effect, no difference, or no relationship** between variables in a study. It represents the default or baseline position that a researcher tries to test against.

### Key points about H₀:

1. **Formulation**: It is usually stated in a way that can be tested statistically.

   * Example: “The new drug has **no effect** on blood pressure” or “There is **no difference** in exam scores between two teaching methods.”

2. **Purpose**:

   * It provides a **reference point** for statistical testing.
   * It allows researchers to quantify the **probability** of observing their data if H₀ were true.

3. **Importance in Hypothesis Testing**:

   * Hypothesis testing is essentially about deciding whether there is enough evidence to **reject H₀** in favor of the **alternative hypothesis (H₁ or Ha)**.
   * Without H₀, we wouldn’t have a clear baseline to compare our data against.
   * It helps control **Type I errors** (false positives), ensuring that any claim of an effect is statistically justified.

**Summary:** H₀ is the “no-change” assumption we test against. It is crucial because it allows researchers to use data to make informed decisions about whether an observed effect is likely due to chance or represents a real phenomenon.

2 What does the significance level (α) represent in hypothesis testing?

  -In hypothesis testing, the **significance level (α)** is a threshold that determines **how strong the evidence must be** before we reject the null hypothesis (H₀).

### Key points about α:

1. **Definition:**

   * α is the **probability of making a Type I error**, which is rejecting H₀ when it is actually true.
   * Common values for α are **0.05, 0.01, or 0.10**, with 0.05 being the most widely used.

2. **Interpretation:**

   * If α = 0.05, it means there is a **5% risk** of concluding that there is an effect when, in reality, there is none.
   * It sets the **cutoff** for the p-value: if the p-value ≤ α, we reject H₀; if p-value > α, we fail to reject H₀.

3. **Importance:**

   * α controls the **strictness** of the test: smaller α → stricter test → less chance of false positives.
   * It balances **risk of error** with **sensitivity** to detect a real effect.

**Example:**

* Suppose a study tests a new drug:

  * H₀: The drug has no effect.
  * α = 0.05.
  * If the p-value from the test is 0.03 → p-value < α → reject H₀ → evidence suggests the drug has an effect.
  * If p-value = 0.08 → p-value > α → fail to reject H₀ → not enough evidence to claim the drug works.

In short, **α is the “risk threshold” for claiming a discovery**—it tells us how much evidence is enough to doubt the null hypothesis.

 3 Differentiate between Type I and Type II errors?

   -In hypothesis testing, **Type I and Type II errors** describe the two kinds of mistakes we can make when making decisions about the null hypothesis (H₀). Here’s a clear breakdown:

---

### **1. Type I Error (False Positive)**

* **Definition:** Rejecting H₀ **when it is actually true**.
* **Symbol:** α (same as the significance level)
* **Meaning:** You think there is an effect or difference, but in reality, there isn’t.
* **Example:**

  * A new drug is tested, but it actually has no effect.
  * The test incorrectly concludes that the drug works.

**Mnemonic:** Type I = “You cried wolf” – seeing an effect that doesn’t exist.

---

### **2. Type II Error (False Negative)**

* **Definition:** Failing to reject H₀ **when it is actually false**.
* **Symbol:** β
* **Meaning:** There really is an effect or difference, but the test misses it.
* **Example:**

  * The same drug actually works, but the study fails to detect it.
  * Concludes that the drug has no effect, even though it does.

**Mnemonic:** Type II = “You missed the wolf” – not seeing an effect that exists.

---

### **Summary Table**

| Error Type | Occurs When…                      | Consequence    | Symbol |
| ---------- | --------------------------------- | -------------- | ------ |
| Type I     | Reject H₀ but H₀ is true          | False positive | α      |
| Type II    | Fail to reject H₀ but H₀ is false | False negative | β      |

---

**Key Insight:**

* **α (significance level)** controls the risk of Type I error.
* **β (power = 1 − β)** relates to Type II error—larger sample sizes usually reduce β.

4.Explain the difference between a one-tailed and two-tailed test. Give an example of each.?

  -In hypothesis testing, the difference between a **one-tailed** and **two-tailed test** lies in **the direction of the effect we are testing for**.

---

### **1. One-Tailed Test**

* **Definition:** Tests for an effect in **only one direction** (either greater than or less than).
* **Use when:** You have a specific expectation about the direction of the effect.
* **Null and Alternative Hypotheses:**

  * H₀: μ ≤ μ₀ (no increase)
  * H₁: μ > μ₀ (increase)
    *or*
  * H₀: μ ≥ μ₀ (no decrease)
  * H₁: μ < μ₀ (decrease)
* **Example:**

  * A company claims their new battery lasts **longer** than the old one.
  * H₀: New battery life ≤ Old battery life
  * H₁: New battery life > Old battery life
  * Only the “longer” direction is of interest.

---

### **2. Two-Tailed Test**

* **Definition:** Tests for an effect in **both directions** (either increase or decrease).
* **Use when:** You are interested in **any difference**, without specifying a direction.
* **Null and Alternative Hypotheses:**

  * H₀: μ = μ₀ (no difference)
  * H₁: μ ≠ μ₀ (there is a difference)
* **Example:**

  * A researcher wants to check if a new teaching method affects exam scores (could increase or decrease).
  * H₀: Mean score = 75
  * H₁: Mean score ≠ 75
  * Both higher and lower scores are of interest.

---

### **Quick Comparison**

| Feature        | One-Tailed Test                 | Two-Tailed Test         |
| -------------- | ------------------------------- | ----------------------- |
| Direction      | Only one (increase or decrease) | Both directions         |
| H₁             | μ > μ₀ or μ < μ₀                | μ ≠ μ₀                  |
| α distribution | All in one tail                 | Split between two tails |
| Example        | Battery life > old battery      | Exam scores ≠ 75        |


 **Tip:**

* Use a **one-tailed test** only if you are confident the effect can only go one way. Otherwise, a **two-tailed test** is safer.

 5.A company claims that the average time to resolve a customer complaint is 10 minutes.

A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.

At α = 0.05, test the claim.

  -Let’s solve this **step by step**. This is a classic **one-sample t-test** problem because the sample size is small (n < 30) and the population standard deviation is unknown.

---

### **Step 1: Define hypotheses**

The company claims the average time is 10 minutes. Our sample shows 12 minutes, so we want to see if the true mean is **different** from 10.

* **Null hypothesis (H₀):** μ = 10
* **Alternative hypothesis (H₁):** μ ≠ 10 (two-tailed test)

---

### **Step 2: Identify given values**

* Sample mean: (\bar{x} = 12)
* Population mean (claimed): μ₀ = 10
* Sample standard deviation: s = 3
* Sample size: n = 9
* Significance level: α = 0.05

---

### **Step 3: Compute the test statistic (t)**

The **t-statistic** formula for a one-sample t-test is:

[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
]

Substitute the values:

[
t = \frac{12 - 10}{3 / \sqrt{9}}
]

Step by step:

1. Compute (\sqrt{n} = \sqrt{9} = 3)
2. Compute (s / \sqrt{n} = 3 / 3 = 1)
3. Compute (t = (12 - 10) / 1 = 2 / 1 = 2)

✅ So, (t = 2)

---

### **Step 4: Find the critical t-value**

* Degrees of freedom: (df = n - 1 = 9 - 1 = 8)
* Two-tailed test at α = 0.05 → split α/2 = 0.025 in each tail

From a t-table or calculator:

[
t_{critical} \approx \pm 2.306
]

---

### **Step 5: Decision rule**

* If (|t| > t_{critical}), **reject H₀**
* If (|t| ≤ t_{critical}), **fail to reject H₀**

Compare:

(|t| = 2 < t_{critical} = 2.306) → **fail to reject H₀**

---

### **Step 6: Conclusion**

At α = 0.05, there is **not enough evidence** to reject the company’s claim that the average complaint resolution time is 10 minutes.

---

💡 **Tip:** Even though the sample mean is 12 minutes, the small sample size and variability mean this difference is **not statistically significant** at 5% level.



6.When should you use a Z-test instead of a t-test?


   -You should use a **Z-test** instead of a **t-test** under certain conditions related to **sample size** and **knowledge of population standard deviation**. Here’s the breakdown:

---

### **1. Z-test Conditions**

Use a Z-test when:

1. **Population standard deviation (σ) is known**

   * The Z-test formula uses σ, which is rarely known in real-world situations.
   * Example: σ = 3 minutes for complaint resolution time.

2. **Sample size is large (n ≥ 30)**

   * By the **Central Limit Theorem**, the sample mean will be approximately normally distributed for large n, even if the population is not perfectly normal.

3. **The data is approximately normally distributed**

   * This is important especially for smaller sample sizes.

---

### **2. T-test Conditions**

Use a t-test when:

1. **Population standard deviation is unknown**

   * Instead, you use the **sample standard deviation (s)** to estimate σ.

2. **Sample size is small (n < 30)**

   * The t-distribution accounts for extra variability due to small samples.

3. **Data is approximately normally distributed**

   * This is more critical for small samples because the t-distribution is sensitive to deviations from normality.

---

### **3. Quick Summary Table**

| Test   | σ known? | Sample size  | Distribution used          |
| ------ | -------- | ------------ | -------------------------- |
| Z-test | Yes      | Large (n≥30) | Standard normal (Z)        |
| T-test | No       | Small or any | t-distribution with df=n−1 |

---

**Example:**

* **Z-test:** You know that the average height of adult men is 175 cm with σ = 7 cm, and you sample 50 men.
* **T-test:** You don’t know σ, and you sample 9 men to estimate the mean height


7.The productivity of 6 employees was measured before and after a training program.?
  -We can solve this using a **paired-sample t-test** because the same employees were measured **before and after** the training.

---

### **Step 1: Define hypotheses**

* **Null hypothesis (H₀):** The training has **no effect** on productivity → mean difference = 0
* **Alternative hypothesis (H₁):** The training **improves productivity** → mean difference > 0 (one-tailed test)

---

### **Step 2: Organize the data**

| Employee | Before | After | Difference (After − Before) |
| -------- | ------ | ----- | --------------------------- |
| 1        | 50     | 55    | 5                           |
| 2        | 60     | 65    | 5                           |
| 3        | 58     | 59    | 1                           |
| 4        | 55     | 58    | 3                           |
| 5        | 62     | 63    | 1                           |
| 6        | 56     | 59    | 3                           |

**Differences (d):** 5, 5, 1, 3, 1, 3

---

### **Step 3: Compute mean and standard deviation of differences**

1. Mean difference ((\bar{d})):

[
\bar{d} = \frac{5 + 5 + 1 + 3 + 1 + 3}{6} = \frac{18}{6} = 3
]

2. Standard deviation of differences (s_d):

First, compute squared deviations from the mean:

[
(d_i - \bar{d})^2
]

* (5 − 3)² = 4
* (5 − 3)² = 4
* (1 − 3)² = 4
* (3 − 3)² = 0
* (1 − 3)² = 4
* (3 − 3)² = 0

Sum of squared deviations = 4 + 4 + 4 + 0 + 4 + 0 = 16

Variance = sum / (n − 1) = 16 / 5 = 3.2

s_d = √3.2 ≈ 1.789

---

### **Step 4: Compute t-statistic**

[
t = \frac{\bar{d}}{s_d / \sqrt{n}}
= \frac{3}{1.789 / \sqrt{6}}
]

Step by step:

1. √n = √6 ≈ 2.449
2. s_d / √n = 1.789 / 2.449 ≈ 0.730
3. t = 3 / 0.730 ≈ 4.11

✅ t ≈ 4.11

---

### **Step 5: Find critical t-value**

* Degrees of freedom: df = n − 1 = 6 − 1 = 5
* One-tailed test at α = 0.05 → t_critical ≈ 2.015

---

### **Step 6: Decision rule**

* If t > t_critical → reject H₀
* Here: t = 4.11 > 2.015 → **reject H₀**

---

### **Step 7: Conclusion**

At α = 0.05, there is **sufficient evidence** to conclude that the training program **improved productivity** of employees.


8.A company wants to test if product preference is independent of gender?

  -We can solve this using a **Chi-square test of independence** because we are testing whether **product preference is independent of gender**.

---

### **Step 1: Organize the data**

From your table, it seems the data is:

|              | Product A | Product B | Row Total |
| ------------ | --------- | --------- | --------- |
| Male         | 30        | 20        | 50        |
| Female       | 10        | 40        | 50        |
| Column Total | 40        | 60        | 100       |

* Total sample size: **N = 100**

---

### **Step 2: State hypotheses**

* **Null hypothesis (H₀):** Product preference is **independent of gender**
* **Alternative hypothesis (H₁):** Product preference is **dependent on gender**

---

### **Step 3: Compute expected counts**

Expected count formula:

[
E_{ij} = \frac{(\text{Row Total}) \times (\text{Column Total})}{\text{Grand Total}}
]

|        | Product A            | Product B            |
| ------ | -------------------- | -------------------- |
| Male   | E = (50×40)/100 = 20 | E = (50×60)/100 = 30 |
| Female | E = (50×40)/100 = 20 | E = (50×60)/100 = 30 |

---

### **Step 4: Compute Chi-square statistic**

Formula:

[
\chi^2 = \sum \frac{(O - E)^2}{E}
]

Compute for each cell:

1. Male, Product A: (30 − 20)² / 20 = 100 / 20 = 5
2. Male, Product B: (20 − 30)² / 30 = 100 / 30 ≈ 3.333
3. Female, Product A: (10 − 20)² / 20 = 100 / 20 = 5
4. Female, Product B: (40 − 30)² / 30 = 100 / 30 ≈ 3.333

Sum: 5 + 3.333 + 5 + 3.333 ≈ 16.666

✅ χ² ≈ 16.67

---

### **Step 5: Degrees of freedom**

[
df = (r - 1) \times (c - 1) = (2-1) \times (2-1) = 1
]

---

### **Step 6: Critical value**

* α = 0.05, df = 1
* From Chi-square table: χ²_critical ≈ 3.841

---

### **Step 7: Decision**

* χ² = 16.67 > 3.841 → **reject H₀**

---

### **Step 8: Conclusion**

At α = 0.05, there is **strong evidence** that **product preference is dependent on gender**.

---

💡 **Tip:** When using a Chi-square test of independence:

1. Organize data in a **contingency table**
2. Compute **expected counts**
3. Apply χ² formula
4. Compare with **critical χ²**
